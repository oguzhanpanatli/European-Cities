```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
# Create a new map with Basemap
map = Basemap(projection='merc', llcrnrlon=-10, llcrnrlat=35, urcrnrlon=30, urcrnrlat=60, resolution='l')

# Draw coastline, countries, and fill the ocean
map.drawcoastlines(linewidth=0.5)
map.drawcountries(linewidth=0.5)
map.fillcontinents(color='lightgray', lake_color='white')

# Draw parallels and meridians
map.drawparallels(range(0, 90, 10), labels=[1, 0, 0, 0], linewidth=0.2, fontsize=8)
map.drawmeridians(range(-180, 180, 20), labels=[0, 0, 0, 1], linewidth=0.2, fontsize=8)

# Define the coordinates of some major European capitals
capitals = {
    'London': (0.1276, 51.5074),
    'Paris': (2.3522, 48.8566),
    'Berlin': (13.4050, 52.5200),
    'Rome': (12.4964, 41.9028),
    'Lisbon': (-9.1393, 38.7223),
    'Vienna': (16.3738, 48.2082),
    'Amsterdam': (4.8952, 52.3702),
    'Budapest': (19.0402, 47.4979),
    'Barcelona': (2.1734, 41.3851),
    'Athens': (23.7275, 37.9838),
}

# Plot the capitals on the map
for capital, (lon, lat) in capitals.items():
    x, y = map(lon, lat)
    plt.plot(x, y, 'ro', markersize=4)
    plt.text(x, y, capital, fontsize=12, color='black',ha='left', va='center')

# Set the title of the plot
plt.xlabel("\nlong")
plt.ylabel("lat\n\n")
# Show the plot
plt.savefig('european_cities_map.png', dpi=600)
plt.show()
```
